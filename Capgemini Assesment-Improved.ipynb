```python
import sys
import pytest

import string
import random
def passgen(length):
    lower = list(string.ascii_lowercase) 
    upper = list(string.ascii_uppercase)  
    special= list(string.punctuation)
    
    
    numupper=3
    numspecial=2
    numd= random.randint(round(0.1*length),round((length-numupper-numspecial)*0.5))
    numlower=length-numupper-numspecial-numd
    digits=[]
    for i in range(numd):
        digits= digits+ [str(random.randint(0,9))]
    
    password=random.choices(lower, k=numlower)+random.choices(upper, k=numupper)+random.choices(special,k=numspecial)+digits

    random.shuffle(password)
    passwordf="".join(map(str, password))
    return passwordf


def test_if_length_is_16_do_we_get_a_password_of_length_16():
    ans = len(passgen(16))
    assert ans == 16


def test_if_length_is_20_do_we_get_a_password_of_length_16():
    ans = len(passgen(20))
    assert ans != 16



if __name__ == "__main__":
    sys.exit(pytest.main([__file__]))
```
